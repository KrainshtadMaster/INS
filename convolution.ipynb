```python
import numpy as np
from scipy.signal import convolve2d
from scipy.signal import convolve
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2, 3, 4, 4,3,2,1])
y = np.array([1, 6, 6, 1])
res = convolve(x,y)

print(x)
print(y)
print(res)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
ax1.stem(x)
ax1.stem(y, linefmt='C3', markerfmt='C3o')
ax2 = fig.add_subplot(122)
ax2.stem(res)
```

```python
import numpy as np

patch = np.array([[1, 1], 
                  [0, 1]])

kernel = np.array([[1, 2], 
                   [3, 4]])

# this is wrong
print(patch.dot(kernel))

# this is how convolution really works
print(patch.flatten()@kernel.flatten())
```
